```python
import pandas as pd
import numpy as np
```

```python
spc_dataset = pd.read_csv('/content/dataset_spc_kaggle.csv',delimiter=',')
```

```python
df_cid = spc_dataset[['customer_products_id','customer_id','customer_state', 'customer_city', 'customer_age', 'customer_avg_income','customer_products_active','product_id','transactions_total_limit','transactions_category']]
df_spc =  spc_dataset[['customer_state', 'customer_city', 'customer_age', 'customer_avg_income','customer_products_active','product_id','transactions_total_limit','transactions_category']]
df_spc.head()
```

```python
df_spc['customer_age'] = df_spc['customer_age'].fillna(df_spc['customer_age'].mean())
```

```python
df_spc['customer_avg_income'] = df_spc['customer_avg_income'].fillna(df_spc['customer_avg_income'].mean())
```

```python
df_spc['customer_products_active'] = df_spc['customer_products_active'].fillna(False)
```

```python
df_spc['transactions_category'] = df_spc['transactions_category'].fillna('Não Especificado')
```

```python
df_spc['transactions_total_limit'] = df_spc['transactions_total_limit'].fillna(df_spc['transactions_total_limit'].mean())
```

```python
df_spc = df_spc.dropna(subset=['customer_state','customer_city','product_id'])
df_cid = df_cid.dropna(subset=['customer_state','customer_city','product_id'])
df_spc.shape
```

```python
df_cid.shape
```

```python
df_spc.head()
```

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
state_encoder = LabelEncoder()
state_encoder.fit(df_spc['customer_state'].unique())
df_spc['customer_state'] = state_encoder.transform(df_spc['customer_state'])
df_spc['customer_state'].unique()
```

```python
city_encoder = LabelEncoder()
city_encoder.fit(df_spc['customer_city'].unique())
df_spc['customer_city'] = city_encoder.transform(df_spc['customer_city'])
df_spc['customer_city'].unique()
```

```python
product_encoder = LabelEncoder()
product_encoder.fit(df_spc['product_id'].unique())
df_spc['product_id'] = product_encoder.transform(df_spc['product_id'])
df_spc['product_id'].unique()
```

```python
category_encoder = LabelEncoder()
category_encoder.fit(df_spc['transactions_category'].unique())
df_spc['transactions_category'] = category_encoder.transform(df_spc['transactions_category'])
df_spc['transactions_category'].unique()
```

```python
df_spc.head()
df_spc.columns
```

```python
scaler = MinMaxScaler()
X = df_spc.drop(columns=['product_id'])
X = scaler.fit_transform(X)
X.shape
```

```python
Y = np.array(df_spc['product_id'])
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
np.unique(y_test)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=19, random_state=0)
yk_train = kmeans.fit_predict(x_train)
yk_test = kmeans.predict(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=20)
model.fit(x_train,yk_train)
```

```python
model.score(x_test, yk_test)
```

```python
def encode_transform(data):
    customer_state = data["customer_state"]
    customer_city = data["customer_city"]
    customer_age = data["customer_age"]
    customer_avg_income = data["customer_avg_income"]
    customer_products_active = data["customer_products_active"]
    transactions_total_limit = data["transactions_total_limit"]
    transactions_category = data["transactions_category"]
    customer_state, = state_encoder.transform([customer_state])
    customer_city, = city_encoder.transform([customer_city])
    transactions_category, = category_encoder.transform([transactions_category])
    scaled_data, = scaler.transform([[
                                     customer_state,
                                     customer_city,
                                     customer_age,
                                     customer_avg_income,
                                     customer_products_active,
                                     transactions_total_limit,
                                     transactions_category
                                    ]])
    return scaled_data

def untransform_product(product):
  product_id, = product_encoder.inverse_transform([product])
  return product_id
```
